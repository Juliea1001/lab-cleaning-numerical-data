```python
def data_type(data):
    data_continious_cols=[]
    data_discrete_cols=[]
    data_categorical_cols=[]
    for x in data.columns:
        if (data[x].dtypes == float or data[x].dtypes == int) and len(data[x].unique()) > 20:
            data_continious_cols.append(x)
           
        elif (data[x].dtypes == float or data[x].dtypes == int) and len(data[x].unique()) <= 20:
            data_discrete_cols.append(x)       
        else:
            data_categorical_cols.append(x)
    data_con= data[data_continious_cols]
    data_dis=data[data_discrete_cols]
    data_cat=data[data_categorical_cols]

        
    return(data_con, data_dis, data_cat)
       
```

```python
data_con, data_dis, data_cat=data_type(data)
```
